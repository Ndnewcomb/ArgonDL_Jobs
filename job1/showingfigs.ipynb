```python
import matplotlib.pyplot as plt
from PIL import Image
import time
import os

def load_and_display_image(image_path):
    try:
        # Open the image file
        img = Image.open(image_path)
        # Display the image
        plt.imshow(img)
        plt.draw()
        plt.pause(1)  # Pause for 1 second
        plt.clf()  # Clear the current figure
    except IOError:
        print("Error in loading image.")

def main(image_path):
    while True:
        load_and_display_image(image_path)
        time.sleep(1)  # Wait for 1 second before loading the next image

if __name__ == "__main__":
    image_path = './PNG/ResourceMonitor.png'  # Replace with your network image path
    main(image_path)
```
